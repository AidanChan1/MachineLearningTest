```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np
import random

def scrambler(slope, intercept): 
    x = np.arange(5,10,0.05)
    y_pred = x.copy()
    # below can be replaced by a lambda function
    for i in range(len(y_pred)):
        y_pred[i] = slope*y_pred[i]+random.uniform(0, 2)+intercept
    return x, y_pred

def least_square(slope,intercept,dataset_x, dataset_y_pred):
    lsr = pow(slope*dataset_x + intercept - dataset_y_pred, 2)
    return np.mean(lsr)
    #a = [0]*len(dataset1)
    #for i in range(len(dataset1)):
    #    a[i] = pow(slope*dataset1[i]+intercept-dataset2[i],2)
    #return sum(a)/len(dataset1)
    #A = a*b
    #B = [number ** 2 for number in a]
    #m = (sum(A)-sum(a)*sum(b))/(sum(b)-pow(sum(a),2))
    #z = (sum(b)-m*sum(a))/len(a)
    #return m, zz8
    #A = np.vstack([a, np.ones(len(a))]).T
    #return np.linalg.lstsq(A, b,rcond=None)[0]
    
x,y_pred = scrambler(random.uniform(3, 8) ,random.uniform(10, 15))


slope = np.arange(3,8,0.05)
intercept = np.arange(10,15,0.05)
#print(slope,"\n", intercept)
another_list = []

for i in intercept:
    lsr_list = []
    for s in slope:
        lsr = least_square(s, i, x, y_pred)
        lsr_list.append(lsr)
    #lsr_array = np.array(lsr_list)
    another_list.append(lsr_list)
    
#print("mmmmmm", another_list)
another_array = np.array(another_list)


#lsr_array.reshape(slope.size, intercept.size)

#test = np.reshape(lsr_array, (-1, 2))

slope, intercept = np.meshgrid(slope, intercept)
#print(slope,"\n", intercept)





#print (test.ndim)
#print (test)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})


# Plot the surface.
surf = ax.plot_surface(slope, intercept, another_array, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
#ax.set_zlim(0, 500)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
#ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## Example for setting up x y z axis

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(-5, 10, 5)
Y = np.arange(-5, 5, 5)
print ("111111", X)
print ("222222",Y)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

print ("333333\n", X)
print ("444444\n", Y)
print (Z)
# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## example for plotting

```python
from matplotlib import cbook
from matplotlib import cm
from matplotlib.colors import LightSource
import matplotlib.pyplot as plt
import numpy as np

# Load and format data
dem = cbook.get_sample_data('jacksboro_fault_dem.npz', np_load=True)
z = dem['elevation']
nrows, ncols = z.shape
x = np.linspace(dem['xmin'], dem['xmax'], ncols)
y = np.linspace(dem['ymin'], dem['ymax'], nrows)
x, y = np.meshgrid(x, y)

region = np.s_[5:50, 5:50]
x, y, z = x[region], y[region], z[region]

# Set up plotu 
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
  
print (type(z))

ls = LightSource(270, 45)
# To use a custom hillshading mode, override the built-in shading and pass
# in the rgb colors of the shaded surface calculated from "shade".
rgb = ls.shade(z, cmap=cm.gist_earth, vert_exag=0.1, blend_mode='soft') 
surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, facecolors=rgb,
                       linewidth=0, antialiased=False, shade=False)

plt.show()
```
